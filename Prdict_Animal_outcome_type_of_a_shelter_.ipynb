```
df_dropped=df[['animal_type','breed','color','outcome_weekday','age_upon_intake_(years)','age_upon_outcome_(years)','sex_upon_intake','outcome_type']]
```

```
df_dropped.shape
```

```
def f(row):
    if row['age_upon_outcome_(years)'] <=2:
        val = 0
    elif row['age_upon_outcome_(years)'] >2:
        val = 1
    else:
        val = -1
    return val
```

```
df_dropped['ChildAdult'] = df_dropped.apply(f, axis=1)
```

```
df_dropped.info()
```

```
from sklearn import preprocessing
# le_anima = preprocessing.LabelEncoder()
# df_dropped.animal_type = le_anima.fit_transform(df_dropped.animal_type)
le_sex = preprocessing.LabelEncoder()
df_dropped.sex_upon_intake = le_sex.fit_transform(df_dropped.sex_upon_intake)
# le_breed = preprocessing.LabelEncoder()
# df_dropped.breed = le_breed.fit_transform(df_dropped.breed)
# le_color = preprocessing.LabelEncoder()
# df_dropped.color = le_color.fit_transform(df_dropped.color)
# le_out = preprocessing.LabelEncoder()
# df_dropped.outcome_type = le_out.fit_transform(df_dropped.outcome_type)
# #Let's see
df_dropped.head()
```

```
df_drop=pd.get_dummies(df_dropped[['animal_type', 'sex_upon_intake','age_upon_outcome_(years)']])
```

```
df_train=df_drop.iloc[:38400,:-1]
df_train_labels=df_dropped[:38400].outcome_type
```

```
df_val=df_drop.iloc[38401:,:-1]
df_val_labels=df_dropped[38401:].outcome_type
```

```
print(df_val.shape,df_val_labels.shape)
```

```
df_train.head()
```

```
df_val.head()
```

```
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score

clf = DecisionTreeClassifier(max_depth=4)

#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
                                                   # random_state=0, stratify=y)

clf.fit(df_train, df_train_labels)
```

```
predictions = clf.predict(df_val)

accuracy_score(df_val_labels, predictions)
```

```
from sklearn.metrics import classification_report

print(classification_report(df_val_labels, predictions))
```

```
from sklearn.ensemble import RandomForestClassifier

rfclf = RandomForestClassifier()
rfclf.fit(df_train, df_train_labels)
```

```
rfclf_predictions = rfclf.predict(df_val)

print(classification_report(df_val_labels, rfclf_predictions))
```

```
from sklearn.metrics import f1_score

f1_score(df_val_labels, rfclf_predictions, average='micro')
```
